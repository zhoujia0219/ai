```python
import random

# 准备数据
x = range(60)
y_shanghai = [random.uniform(15,18) for i in x]
# 增加北京的温度数据
y_beijing = [random.uniform(1, 3) for i in x]

# 创建画布
# plt.figure(figsize=(20,8), dpi=100)
# 改面向对象 画布对象 绘图区对象
figure,axes = plt.subplots(nrows=1,ncols=2,figsize=(20,8), dpi=100)

# 绘制折线图
# plt.plot(x,y)
# plt.plot(x, y_shanghai, label="上海")
# 使用多次plot可以画多个折线
# plt.plot(x, y_beijing, color='r', linestyle='--', label="北京")
axes[0].plot(x, y_shanghai, label="上海")
axes[1].plot(x, y_beijing, color='r', linestyle='--', label="北京")

# 显示图例
# plt.legend(loc = "best")
axes[0].legend()
axes[1].legend()


# 添加网格显示
# plt.grid(True, linestyle="--", alpha=0.5)
axes[0].grid(True, linestyle="--", alpha=0.5)
axes[1].grid(True, linestyle="--", alpha=0.5)

# 添加描述信息
# plt.title("中午11点到12点的温度变化状况")
# plt.xlabel("时间")
# plt.ylabel("温度")
axes[0].set_title("上海中午11点到12点的温度变化状况")
axes[0].set_xlabel("时间")
axes[0].set_ylabel("温度")
axes[1].set_title("北京中午11点到12点的温度变化状况")
axes[1].set_xlabel("时间")
axes[1].set_ylabel("温度")

# 添加自定义刻度
# 准备x轴刻度中文描述
xticks_label = ["11点{}分".format(i) for i in x[::5]]
# plt.xticks(x[::5], xticks_label)
# plt.yticks(range(40)[::5])
axes[0].set_xticks(x[::5], xticks_label)
axes[0].set_yticks(range(40)[::5])
axes[1].set_xticks(x[::5], xticks_label)
axes[1].set_yticks(range(40)[::5])

# 显示图像
plt.show()
```

##  柱状图 电影票房收入对比

```python
# 1) 准备数据
# 电影名字
movie_name = ['雷神3：诸神黄昏','正义联盟','东方快车谋杀案','寻梦环游记','全球风暴','降魔传','追捕','七十七天','密战','狂兽','其它']
# 横坐标
x = range(len(movie_name))
# 票房数据
y = [73853,57767,22354,15969,14839,8725,8716,8318,7916,6764,52222]

# 2) 创建画布
plt.figure(figsize=(20,8), dpi=100)

# 3) 绘制柱状图
plt.bar(x, y, width=0.5, color=['b','r','g','y','c','m','y','k','c','g','b'])

# 修改x轴刻度显示
plt.xticks(x,movie_name)

# 添加网格显示
plt.grid(linestyle="--", alpha=0.5)

# 添加标题
plt.title("电影票房收入对比")

# 4) 显示图像
plt.show()
```

### 比较相同天数的票房

```python
# 1) 准备数据
# 电影名字
movie_name = ['雷神3：诸神黄昏','正义联盟','寻梦环游记']

first_day = [10587.6,10062.5,1275.7]
first_weekend=[36224.9,34479.6,11830]

# 横坐标
x = range(len(movie_name))

# 2) 创建画布
plt.figure(figsize=(20,8), dpi=100)

# 3) 绘制柱状图
plt.bar(x, first_day, width=0.2, label="首日票房")
plt.bar([i+0.2 for i in x], first_weekend, width=0.2, label="首周票房")

# 修改x轴刻度显示
plt.xticks([i+0.1 for i in x],movie_name)

# 添加网格显示
plt.grid(linestyle="--", alpha=0.5)

# 显示图例
plt.legend()

# 添加标题
plt.title("电影票房收入对比")

# 4) 显示图像
plt.show()
```
