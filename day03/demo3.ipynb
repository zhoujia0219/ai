```python
data = pd.read_csv("./stock_day/stock_day.csv")
```

```python
data = data.drop(["ma5","ma10","ma20","v_ma5","v_ma10","v_ma20"], axis=1)
```

```python
data
```

```python
# 直接使用行列索引 先列后行
data["open"]["2018-02-06"]
```

```python
# 使用索引名获取
data.loc["2018-02-06"]["open"]
```

```python
data.loc["2018-02-06", "open"]
```

```python
# 使用数字索引获取
data.iloc[1,0]
```

```python
# 组合索引 ix 
data.head()
```

```python
data.ix[2,"high"]
```

```python
data["volume"] = 1000
```

```python
data.volume = 1200
```

```python
data
```

```python
# 对DataFrame的内容进行排序
data.sort_values(by="open", ascending=False).head()
```

```python
# 值相等 多值排序
data.sort_values(by=["open", "high"], ascending=False).head()
```

```python
# 对索引进行排序
data.sort_index().head()
```

```python
# Series 对内容进行排序
sr = data["open"]
```

```python
sr.sort_values(ascending=False).head()
```

```python
# 对索引进行排序
sr.sort_index().head()
```

```python
data["close"].sub(data["open"]).head()
```

```python
# 逻辑运算
data["p_change"].head() > 2
```

```python
# 布尔索引
data[data.p_change > 2].head()
```

```python
data[(data.p_change > 2) & (data.p_change < 15)].head()
```

```python
# 逻辑运算函数
# 更简洁的查询方法
data.query("p_change > 2 & open > 15").head()
```

```python
data[data.turnover.isin([4.19, 2.39])]
```

```python
data.idxmax(axis=0)
```

```python
# 累计统计函数
data["p_change"].cumsum().plot()
```

```python
# 自定义运算
# 定义一个对列，最大值-最小值的函数
data[["open", "close"]].apply(lambda x: x.max()-x.min(), axis=0)
```
